```python
dataset = pd.read_csv('DigitalAd_dataset.csv')
print(dataset)
```

**Summarize dataset**

```python
print(dataset.shape)
print(dataset.head(10))
```

**Segregate Dataset into X(Input) & Y(Output)**

```python
X = dataset.iloc[:,:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

**Splitting dataset into Train & Test**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.25 , random_state = 0)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

**Training the Model**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

**Predicting, Whether the New Customer with Age & Salary will Buy the Product or Not**

```python
age = int(input("Enter the New Customer's Age : "))
sal = int(input("Enter the New Customer's Salary : "))
newCust = [[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result == 1:
  print("Customer will Buy the Product")
else:
  print("Customer Will Not Buy the Product")
```

**Prediction for all Test Data**

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model : {0}% ".format(accuracy_score(Y_test,Y_pred)*100))
```
